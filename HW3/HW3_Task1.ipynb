```python
def calculate_centroid(points):
    return np.average(points, axis = 0)
```

```python
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

def cosine_distance(p1, p2):
    return 1 - (np.dot(p1, p2) / np.sqrt(np.dot(p1, p1) * np.dot(p2, p2)))

def jaccard_distance(p1, p2):
    return 1 - (np.sum(np.minimum(p1, p2)) / np.sum(np.maximum(p1, p2)))
```

```python
def k_means(data, k, initial_centroids, distance):
    centroids = initial_centroids
    clusters_index = None

    while True:
        distances = [[distance(x, centroid) for centroid in centroids] for x in data]
        clusters_index = np.argmin(distances, axis = 1)

        new_centroids = [np.mean(data[clusters_index == i], axis = 0) for i in range(k)]

        optimized = True
        for i in range(k):
            for j in range(len(centroids[i])):
                if centroids[i][j] != new_centroids[i][j]:
                    optimized = False

        if optimized:
            break

        centroids = new_centroids

    return centroids, clusters_index
```

```python
k = len(labels[0].unique())
```

```python
data_arr = data.to_numpy()
```

```python
labels_arr = [labels.to_numpy()[i][0] for i in range(labels.shape[0])]
```

```python
# Initialize k random points as initial centroids
initial_centroids = []
for _ in range(k):
    i = np.random.randint(data_arr.shape[0])
    initial_centroids.append(data_arr[i])
```

```python
euclidean_centroids, euclidean_clusters_index = k_means(data_arr, k, initial_centroids, euclidean_distance)
cosine_centroids, cosine_clusters_index = k_means(data_arr, k, initial_centroids, cosine_distance)
jaccard_centroids, jaccard_clusters_index = k_means(data_arr, k, initial_centroids, jaccard_distance)
```

#### Q1: Compare the SSEs of Euclidean-K-means, Cosine-K-means, Jarcard-K-means

```python
def sum_of_squared_error(points, centroids, clusters_index):
    sse = 0
    for i in range(len(points)):
        sse += np.sum((centroids[clusters_index[i]] - points[i]) ** 2)

    return sse
```

```python
sse_df = pd.DataFrame({
    "Distance_Metric": ["Euclidean", "Cosine", "Jaccard"],
    "SSE": [
        sum_of_squared_error(data_arr, euclidean_centroids, euclidean_clusters_index),
        sum_of_squared_error(data_arr, cosine_centroids, cosine_clusters_index),
        sum_of_squared_error(data_arr, jaccard_centroids, jaccard_clusters_index)
    ]
})
sse_df = sse_df.sort_values(by = ["SSE"])
sse_df
```

#### Q2: Compare the accuracies of Euclidean-K-means Cosine-K-means, Jarcard-K-means.

```python
def predict_cluster_labels(labels, k, clusters_index):
    cluster_index_label = [[0 for _ in range(k)] for _ in range(k)]

    for i in range(len(clusters_index)):
        cluster_index_label[clusters_index[i]][labels[i]] += 1

    predicted_cluster_labels = np.argmax(cluster_index_label, axis = 1)

    predicted_labels = []
    for i in range(len(clusters_index)):
        predicted_labels.append(predicted_cluster_labels[clusters_index[i]])

    return predicted_labels
```

```python
def compute_accuracy(predicted_labels, actual_labels):
    return np.average(predicted_labels == np.array(actual_labels)) * 100
```

```python
euclidean_predicted_labels = predict_cluster_labels(labels_arr, k, euclidean_clusters_index)
cosine_predicted_labels = predict_cluster_labels(labels_arr, k, cosine_clusters_index)
jaccard_predicted_labels = predict_cluster_labels(labels_arr, k, jaccard_clusters_index)
```

```python
accuracy_df = pd.DataFrame({
    "Distance_Metric": ["Euclidean", "Cosine", "Jaccard"],
    "Accuracy": [
        compute_accuracy(euclidean_predicted_labels, labels_arr),
        compute_accuracy(cosine_predicted_labels, labels_arr),
        compute_accuracy(jaccard_predicted_labels, labels_arr)
    ]
})
accuracy_df = accuracy_df.sort_values(by = ["Accuracy"], ascending = False)
accuracy_df
```

#### Q3: Set up the same stop criteria: “when there is no change in centroid position OR when the SSE value increases in the next iteration OR when the maximum preset value (e.g., 500, you can set the preset value by yourself) of iteration is complete”, for Euclidean-K-means, Cosine-Kmeans, Jarcard-K-means. Which method requires more iterations and times to converge?

```python
def k_means_no_change_in_centroid(data, k, initial_centroids, distance):
    start_time = time()

    centroids = initial_centroids
    no_of_iterations = 0
    sse = float("inf")

    while True:
        no_of_iterations += 1
        distances = [[distance(x, centroid) for centroid in centroids] for x in data]
        clusters_index = np.argmin(distances, axis = 1)

        new_centroids = [np.mean(data[clusters_index == i], axis = 0) for i in range(k)]
        new_sse = sum_of_squared_error(data, new_centroids, clusters_index)

        optimized = True
        for i in range(k):
            for j in range(len(centroids[i])):
                if centroids[i][j] != new_centroids[i][j]:
                    optimized = False

        if optimized:
            break

        centroids = new_centroids
        sse = new_sse

    end_time = time()

    return no_of_iterations, sse, end_time - start_time
```

```python
def k_means_increase_in_sse(data, k, initial_centroids, distance):
    start_time = time()

    centroids = initial_centroids
    no_of_iterations = 0
    sse = float("inf")

    while True:
        no_of_iterations += 1
        distances = [[distance(x, centroid) for centroid in centroids] for x in data]
        clusters_index = np.argmin(distances, axis = 1)

        new_centroids = [np.mean(data[clusters_index == i], axis = 0) for i in range(k)]
        new_sse = sum_of_squared_error(data, new_centroids, clusters_index)

        if new_sse >= sse:
            break

        centroids = new_centroids
        sse = new_sse

    end_time = time()

    return no_of_iterations, sse, end_time - start_time
```

```python
def k_means_maximum_preset_value(data, k, initial_centroids, distance):
    start_time = time()

    centroids = initial_centroids
    no_of_iterations = 0
    sse = float("inf")

    while no_of_iterations < 500:
        no_of_iterations += 1
        distances = [[distance(x, centroid) for centroid in centroids] for x in data]
        clusters_index = np.argmin(distances, axis = 1)

        new_centroids = [np.mean(data[clusters_index == i], axis = 0) for i in range(k)]
        new_sse = sum_of_squared_error(data, new_centroids, clusters_index)

        centroids = new_centroids
        sse = new_sse

    end_time = time()

    return no_of_iterations, sse, end_time - start_time
```

```python
euclidean_centroid_based_no_of_iterations, euclidean_centroid_based_sse, euclidean_centroid_based_time = k_means_no_change_in_centroid(data_arr, k, initial_centroids, euclidean_distance)
euclidean_sse_based_no_of_iterations, euclidean_sse_based_sse, euclidean_sse_based_time = k_means_increase_in_sse(data_arr, k, initial_centroids, euclidean_distance)
euclidean_preset_based_no_of_iterations, euclidean_preset_based_sse, euclidean_preset_based_time = k_means_maximum_preset_value(data_arr, k, initial_centroids, euclidean_distance)
```

```python
euclidean_df = pd.DataFrame({
    "Stop_Criteria": ["No Change in Centroid", "Increase in SSE", "Maximum preset value"],
    "No_of_Iterations": [euclidean_centroid_based_no_of_iterations, euclidean_sse_based_no_of_iterations, euclidean_preset_based_no_of_iterations],
    "Time_Taken": [euclidean_centroid_based_time, euclidean_sse_based_time, euclidean_preset_based_time],
    "SSE": [euclidean_centroid_based_sse, euclidean_sse_based_sse, euclidean_preset_based_sse]
})
euclidean_df[["Stop_Criteria", "No_of_Iterations", "Time_Taken"]].sort_values(by = ["No_of_Iterations", "Time_Taken"])
```
